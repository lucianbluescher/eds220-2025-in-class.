## Selecting a single column

Simplest case: selecting a single column by column name.

General syntax:

'''
df['column-name']
'''

Label based subsetting. Means we select data from our data frame using the *names* of the columns *not their position*.

### Example


````python
# Select number of bird species observed at Mugu Lagoon in spring
mul_spring = df["MUL_spring"]

mul_spring
````

````python
type(mul_spring)
````

We can do label-based subsetting of a single column using attribute syntax

'df.column_name'

### Example

````python
df.MUL_spring
````

Favor 'df['column_name']' instead of 'df.column_name'

Why? Great answers!

## Selecting multiple columns...

... using a list of column names. 

Syntax:

'''
df[  [column_1', 'column_10', 'column_7']  ]
Notice: there are double square brackets! This is because we are passing a list of names '['column_1', 'column_2'] to the selection brackets []

### Example

````python
# Select species aboundance in Tijuana Estuary during winter and fall
tje_wf = df[['TJE_winter', 'TJE_fall']]
tje_wf
````

````python
# Check the type of tje_wf
print(type(tje_wf))
# Check the shape of the selection 
print(tje_wf.shape)
````

## ... using a slice

To select a slice of the columns we will use a special case of 'loc' selection.

Syntax:
'''
df.loc[ :, 'column_start':'column_end']

'column_start' and 'column_end' = starting and ending point of column slice, slice includes both endpoints.

- the first value passed to 'loc' is used to select rows, using a ':' as the row-selection parameter means "select all the rows"

### Example

````python
# Select columns between CSM_winter and MUL_fall
csm_mul = df.loc[ : , 'CSM_winter':'MUL_fall']
csm_mul
````

## Selecting rows...

### ...using a condtion

Syntax:
'''
df[condition_on_rows]
'''

That condition_on_rows can be many things.

### Example

We are interesed in all data after 2020

````python
# Select all rows with year >2020

post_2020 = df[df['year']>2020]
post_2020
````

condition for our rows = 'df['year']>2020' this is a 'pandas.Series' with **boolean values** (True or False)
indicates which rows satisfy the confition year>2020

````python
# Check the type of df['year']>2020
print(type(df['year']>2020))

df['year']>2020
````

### Check-on
Get the subset of the data frame on which the San Dieguito Wetland has at least 75 species recorded during spring

````python
sdw_spring = df[df['SDW_spring']>=75]
sdw_spring
````

### Example

Suppose we want to look at data from 2012 to 2015 (including both years).

````python
subset = df[df['year'].between(2012,2015)]
subset
````

- `df['year']` = column with year values, this a 'pandas.Series'
- 'df['year'].between()' we have that 'between()' is a method for the pandas.Series. Calling it using '.'
- (2012,2015): parameters. If we look at the documentation, we can see that the method includes both endpoints. 
- 'df['year'].between(2012,2015)' = a pandas.Series of boolean values indicating which rows have year equal to 2012-2015
- when you put `df['year'].between(2012,2015)` in the selection brackets, we obtain the rows we need

## Avoid using `loc` for selecting only rows. 
It is equvilant to write

```
# Select rows with year<2015
df[df['year']<2015]
```
and
```
df.loc[df['year']<2015 , : ]
```

In the second one:
- using `df.loc[ row-selection, column-selection]`
- the row selection is the condition `df['year']<2015]`
- the column selection is `:` = all column

## ... using multiple conditions

We can combine conditions to select rows by surrounding each in parenthesis `()` and usuing the or operator `|` and the and operator `&`.

### Example :or

````python
# Select rows in CSM with more than 50 species in winter or fall
df[ (df['CSM_winter']>50) | (df['CSM_fall']>50) ]
````

Select rows in which both CSM and SDW have more than 60 bird species during spring

````python
df[(df['CSM_spring']>60) & (df['SDW_spring']>60)]
````

````python
cond1 = (df['CSM_winter']>50)
cond2 = (df['CSM_fall']>50)

df[cond1 | cond2]
````

### Selecting rows by position

Sometimes we may want to select rows based on their *actual position* in the data frame

Syntax:
'''
df.iloc[row-indices]
'''
iloc = integer based indexing

````python
df.iloc[4]
````

````python
df2 = df.set_index('year')
df2
````

````python
# Selecting by 5th row index // loc is indexing by LABELS
df2.loc[2014]
````

````python
df2.iloc[4]
````

## Select rows and columns simultaneously

Can be done using 'loc' (labels) or 'iloc' (positions)

### ... by labels or conditions

Syntax"
```
df.loc[ row-selection, column-selection]
```

### Examples
Select winter surveys for mugu lagoon and tigjuanan estuary after 2020
Using `loc`

````python
df.loc[df['year']>2020, ['MUL_winter', 'TJE_winter']]
````

### ... by position
Use `iloc`
Synatx:
```
df.iloc[ row-indices , column-indices ]
```
### Example
Select rows 3-7 including 7 and fourth and fifth columns

````python
df.iloc[3:7, [3,4]]
````
